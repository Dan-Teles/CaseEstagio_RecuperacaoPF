```python
import pandas as pd

# Carregar base (use caminho absoluto ou faça upload do arquivo)
df_raw = pd.read_csv("/snapshot_carteira_raw.csv")

# Indicadores iniciais
snapshot_rows_raw = len(df_raw)
snapshot_dupes_raw = df_raw.duplicated(subset=["id_contrato","mes_ref"]).sum()
snapshot_neg_dias_raw = (df_raw["atraso"] < 0).sum()
snapshot_neg_saldo_raw = (df_raw["saldo"] < 0).sum()
em_dia_com_saldo_pos = ((df_raw["atraso"] == 0) & (df_raw["saldo"] > 0)).sum()

# Tratamento duplicidades: manter maior saldo
df_clean = df_raw.sort_values("saldo", ascending=False).drop_duplicates(subset=["id_contrato","mes_ref"])

# Corrigir valores inválidos
df_clean["atraso"] = df_clean["atraso"].clip(lower=0)
df_clean["saldo"] = df_clean["saldo"].clip(lower=0)

# Indicadores pós-limpeza
snapshot_rows_clean = len(df_clean)
snapshot_dupes_clean = df_clean.duplicated(subset=["id_contrato","mes_ref"]).sum()

# Relatório final
report = pd.DataFrame({
    "snapshot_rows_raw":[snapshot_rows_raw],
    "snapshot_rows_clean":[snapshot_rows_clean],
    "snapshot_dupes_raw":[snapshot_dupes_raw],
    "snapshot_dupes_clean":[snapshot_dupes_clean],
    "snapshot_neg_dias_raw":[snapshot_neg_dias_raw],
    "snapshot_neg_saldo_raw":[snapshot_neg_saldo_raw],
    "em_dia_com_saldo_pos":[em_dia_com_saldo_pos]
})

report.to_csv(r"C:\Users\danie\OneDrive\Desktop\Case Itau\Case - Estágio em Dados - Itaú\dq_report.csv", index=False)

# Evidências
print("Antes:", df_raw.head())
print("Depois:", df_clean.head())
```
